```python
import pandas as pd
import matplotlib.pyplot as plt

#loading the dataset
data = pd.read_csv('dataset.csv')

#making the bar chart
data['quality'].value_counts().sort_index().plot(kind='bar',edgecolor='black')
plt.xlabel('Quality Score')
plt.ylabel('Count of Wines')
plt.title('Count of wines for each quality score')
plt.show()
```

```python
# Scatter plot for alcohol vs. pH
plt.scatter(data['alcohol'], data['pH'])

# customise scatterplot
plt.title('Alcohol vs. pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()

# Checking correlation
correlation = data['alcohol'].corr(data['pH'])
print(f'Correlation between alcohol and pH:', correlation)
```

```python
#make the histogram
bins = [8,9,10,11,12,13,14,15]
plt.hist(data['alcohol'],bins=bins,edgecolor='black')

# customise the histogramS
plt.title('Alcohol Distribution')
plt.xlabel('Alcohol')
plt.ylabel('Freqency of Alcohol')
plt.show()
```

```python
# create a box plot
data.boxplot(column='residual sugar', by='quality', grid=True)

# customize the plot
plt.title('Spread of Residual Sugar Values by Quality Score')
plt.suptitle('')  # Removes the automatic 'Boxplot grouped by quality' title
plt.xlabel('Quality Score')
plt.ylabel('Residual Sugar')
plt.show()
```

```python
table = data.corr()
table
```

```python
import seaborn as sns
plt.figure(figsize=(15,8))
sns.heatmap(table,cmap='RdBu',annot=True)
```

```python
# create line plot
plt.plot(data['free sulfur dioxide'],data['total sulfur dioxide'])

# customise line plot
plt.title('Free vs Total Sulfur Dioxide')
plt.xlabel('Free Sulfur Dioxide')
plt.ylabel('Total Sulfer Dioxide')
plt.show()
```

```python

#Mean value of the Citric acid column
meanvalue = data['citric acid'].mean()
print('The mean value of Citric acid column is', meanvalue)
```

```python
# Get minimum values for each numerical column
min_values = data.min()

# Get maximum values for each numerical column
max_values = data.max()

# Display the results
print("Minimum values:\n", min_values)
print("\nMaximum values:\n", max_values)
```

```python
#Filter wines with a pH greater than 3.5 and alcohol content greater than 10.
filtered_wines = data[(data['pH']>3.5) & (data['alcohol']>10)]

#Calculate the average quality of these wines
average_wines = filtered_wines['quality'].mean()
print("The average quality of wines with pH > 3.5 and alcohol content > 10 is",average_wines)
```

```python
#Create a new column of acidity ratio
acidity_ratio = data['fixed acidity'] / data['volatile acidity']

#Mean of acidity ratio
acidity_ratio_mean = acidity_ratio.mean()
print('The mean of Acidity ratio is',acidity_ratio_mean)
```
